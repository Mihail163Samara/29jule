Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170. Данная промежуточная аттестация оценивается по системе "зачет" / "не зачет". "Зачет" ставится, если Слушатель успешно выполнил задание. "Незачет" ставится, если Слушатель не выполнил задание. Критерии оценивания: 1 - Слушатель провел дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов.

Для проведения дисперсионного анализа (ANOVA) для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов, выполним следующие шаги:

1. Соберем данные роста спортсменов по группам.

2. Проведем однофакторный дисперсионный анализ ANOVA.

3. Оценим результаты анализа.

```python
import scipy.stats as stats

# Данные
football_players = [173, 175, 180, 178, 177, 185, 183, 182]
hockey_players = [177, 179, 180, 188, 177, 172, 171, 184, 180]
weightlifters = [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]

# Проведение однофакторного дисперсионного анализа
f_statistic, p_value = stats.f_oneway(football_players, hockey_players, weightlifters)

f_statistic, p_value
```

# **Результаты анализа**

По результатам однофакторного дисперсионного анализа ANOVA мы получили следующие значения:  

F-статистика: 5.500.

p-значение: 0.010.

**Интерпретация результатов**.

Значение F-статистики (5.500) указывает на наличие различий между группами. Чем больше значение F, тем больше различий между средними значениями групп.  


p-значение (0.010) указывает на вероятность того, что наблюдаемые различия в средних значениях могли возникнуть случайно. Обычно в статистике используют уровень значимости (α) 0.05.  


**Вывод**

Поскольку полученное p-значение (0.010) меньше уровня значимости 0.05, мы отвергаем нулевую гипотезу о том, что средние значения роста в трех группах (футболисты, хоккеисты и штангисты) равны. Это означает, что есть статистически значимые различия в среднем росте среди этих групп спортсменов.
